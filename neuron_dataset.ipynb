### Load necessary libraries

```python
import pandas as pd
import numpy as np
import warnings
import psycopg2
import openpyxl
```

```python
# 1. Shrinkage dataset
# Loading the Excel file
excel_file = 'neuron_dataset_shrinkage.xlsx'

# Read all sheets into a dictionary of DataFrames
excel_data = pd.read_excel(excel_file, sheet_name=None, engine='openpyxl')

# Iterate through each sheet and save it as a separate CSV file
for sheet_name, data in excel_data.items():
   
    csv_file = f'{sheet_name}.csv'  
    data.to_csv(csv_file, index=False)
    print(f'Saved {sheet_name} as {csv_file}')
```

```python
# Summary dataset
# Load the Excel file
excel_file = 'neuron_dataset_summary.xlsx' 

# Read all sheets into a dictionary of DataFrames
excel_data = pd.read_excel(excel_file, sheet_name=None, engine='openpyxl')

# Iterate through each sheet and save it as a separate CSV file
for sheet_name, data in excel_data.items():
    
    csv_file = f'{sheet_name}.csv'  # The file name will be the sheet name
    data.to_csv(csv_file, index=False)
    print(f'Saved {sheet_name} as {csv_file}')
```

### Transform bouton_number to unique identifiers in 'vesicle' csv files

```python
CB_den1_before_shrink_vesicle = pd.read_csv('CB_den1_before_shrink_vesicle.csv')

df = CB_den1_before_shrink_vesicle

# Function to add sequential labels
def add_labels(group):
    return [f"{group.name}_{i+1}" for i in range(len(group))]

# Apply the function to each group
df['bouton_number'] = df.groupby('bouton_number')['bouton_number'].transform(add_labels)
df
```

```python
CB_den1_after_shrink_vesicle = pd.read_csv('CB_den1_after_shrink_vesicle.csv')

df2 = CB_den1_after_shrink_vesicle

# Function to add sequential labels
def add_labels(group):
    return [f"{group.name}_{i+1}" for i in range(len(group))]

# Apply the function to each group
df2['bouton_number'] = df2.groupby('bouton_number')['bouton_number'].transform(add_labels)

df2
```

```python
CB_den2_before_shrink_vesicle = pd.read_csv('CB_den2_before_shrink_vesicle.csv')

df3 = CB_den2_before_shrink_vesicle

# Function to add sequential labels
def add_labels(group):
    return [f"{group.name}_{i+1}" for i in range(len(group))]

# Apply the function to each group
df3['bouton_number'] = df3.groupby('bouton_number')['bouton_number'].transform(add_labels)

df3
```

```python
CB_den2_after_shrink_vesicle = pd.read_csv('CB_den2_after_shrink_vesicle.csv')

df4 = CB_den2_after_shrink_vesicle

# Function to add sequential labels
def add_labels(group):
    return [f"{group.name}_{i+1}" for i in range(len(group))]

# Apply the function to each group
df4['bouton_number'] = df4.groupby('bouton_number')['bouton_number'].transform(add_labels)

df4
```

```python
df.to_csv('CB_den1_before_shrink_vesicle.csv', index=False)
df2.to_csv('CB_den1_after_shrink_vesicle.csv', index=False)
df3.to_csv('CB_den2_before_shrink_vesicle.csv', index=False)
df4.to_csv('CB_den2_after_shrink_vesicle.csv', index=False)
```
