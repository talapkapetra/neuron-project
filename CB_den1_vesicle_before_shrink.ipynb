### Load Necessary Libraries

```python
import pandas as pd
import numpy as np
import warnings
```

```python
# Loading data from the xlsx file into a DataFrame and save in a variable named CB_den1.
CB_den1 = pd.read_excel('CB_den1_before_shrink.xlsx')
```

```python
# Centroid_values have been converted to string data type.
CB_den1['centroid_values']=CB_den1['centroid_values'].astype('string')
```

```python
CB_den1.info()
```

```python
CB_den1_clean = CB_den1.dropna().reset_index(drop=True)
CB_den1_clean
```

```python
CB_den1_clean.info()
```

```python
data = CB_den1_clean 

# Function to transform the centroid values
def transform_centroid(centroid_str):
    # Remove the parentheses and split by comma
    coords = centroid_str.strip('()').split(', ')
    
    # Convert the values from 'x,y' format to 'x.y'
    formatted_coords = [coord.replace(',', '.') for coord in coords]
    
    # Join the coordinates back together in the desired format
    return f"({', '.join(formatted_coords)})"

# Apply the transformation to the 'centroid_values' column and store it in a new column
CB_den1_clean['centroid_values_new'] = CB_den1_clean['centroid_values'].apply(transform_centroid)

CB_den1_clean
```

```python
data = CB_den1_clean

# Function to remove the third value from the centroid values
def remove_third_value(centroid_str):
    # Remove parentheses and split by comma
    coords = centroid_str.strip('()').split(', ')
    if len(coords) >= 3:
        # Keep only the first two values
        return f"({coords[0]}, {coords[1]})"
    else:
        # Return the string as is if it doesn't have the expected format
        return centroid_str

# Apply the function to modify the relevant columns
CB_den1_clean['centroid_values'] = CB_den1_clean['centroid_values'].apply(remove_third_value)
CB_den1_clean['centroid_values_new'] = CB_den1_clean['centroid_values_new'].apply(remove_third_value)
warnings.filterwarnings('ignore')

CB_den1_clean
```

```python
data = CB_den1_clean

# Extract the x and y coordinates from 'centroid_values_new'
CB_den1_clean[['x_coord', 'y_coord']] = CB_den1_clean['centroid_values_new'].str.extract(r'\(([^,]+),\s*([^)]+)\)')

# Convert the extracted values to float
CB_den1_clean['x_coord'] = CB_den1_clean['x_coord'].astype(float)
CB_den1_clean['y_coord'] = CB_den1_clean['y_coord'].astype(float)

CB_den1_clean
```

```python
data = CB_den1_clean

# Function to calculate Manhattan distance
def manhattan_distance(point1, point2):
    return abs(point1[0] - point2[0]) + abs(point1[1] - point2[1])

# Function to calculate Euclidean distance
def euclidean_distance(point1, point2):
    return np.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)

# Initialize a list to store the results
results = []

# Loop through each bouton_number group
for bouton_number, group in CB_den1_clean.groupby('bouton_number'):
    for index, row in group.iterrows():
        original_point = (row['x_coord'], row['y_coord'])
        
        # Calculate Manhattan distances to all other points in the group, excluding the original point
        distances = group.apply(
            lambda r: manhattan_distance(original_point, (r['x_coord'], r['y_coord'])) if r.name != index else np.inf, 
            axis=1
        )
         # Identify the index of the closest point based on Manhattan distance
        closest_point_index = distances.idxmin()
        closest_point = (group.loc[closest_point_index, 'x_coord'], group.loc[closest_point_index, 'y_coord'])
        
        # Calculate the Euclidean distance between the original point and the closest point
        euclidean_dist = euclidean_distance(original_point, closest_point)
        
        # Store the results (you can modify what you want to store)
        results.append({
            'bouton_number': bouton_number,
            'original_point': original_point,
            'closest_point': closest_point,
            'euclidean_distance': euclidean_dist
        })

# Convert results to a DataFrame
CB_den1_clean_results_df = pd.DataFrame(results)

# Display the results
CB_den1_clean_results_df
```

```python
CB_den1_clean_results_df['euclidean_distance_corr_nanometer'] = CB_den1_clean_results_df['euclidean_distance'] * 10
CB_den1_clean_results_df
```

```python
CB_den1_clean['bouton_number'] = CB_den1_clean['bouton_number'].astype('string')
CB_den1_clean['bouton_number'] = CB_den1_clean['bouton_number'].str.replace('b21_ss6', 'b21')
CB_den1_clean['bouton_number'] = CB_den1_clean['bouton_number'].str.replace('b21_ss7', 'b21')
CB_den1_clean['bouton_number'] = CB_den1_clean['bouton_number'].str.replace('b', '').astype(int)

CB_den1_clean_sorted = CB_den1_clean.sort_values(by='bouton_number', ascending=True)
CB_den1_clean_sorted = CB_den1_clean_sorted.reset_index(drop=True)

CB_den1_clean_sorted
```

```python
CB_den1_clean_results_df['bouton_number'] = CB_den1_clean_results_df['bouton_number'].astype('string')
CB_den1_clean_results_df['bouton_number'] = CB_den1_clean_results_df['bouton_number'].str.replace('b21_ss6', 'b21')
CB_den1_clean_results_df['bouton_number'] = CB_den1_clean_results_df['bouton_number'].str.replace('b21_ss7', 'b21')
CB_den1_clean_results_df['bouton_number'] = CB_den1_clean_results_df['bouton_number'].str.replace('b', '').astype(int)

CB_den1_clean_results_df_sorted = CB_den1_clean_results_df.sort_values(by='bouton_number', ascending=True)
CB_den1_clean_results_df_sorted = CB_den1_clean_results_df_sorted.reset_index(drop=True)

CB_den1_clean_results_df_sorted
```

```python
# Select 2 columns from each DataFrame
CB_den1_clean_sorted_selected = CB_den1_clean_sorted[['bouton_number', 'synapse_type', 'x_coord', 'y_coord']]  # Choose columns from the first DataFrame
CB_den1_clean_results_df_sorted_selected = CB_den1_clean_results_df_sorted[['original_point', 'closest_point', 'euclidean_distance_corr_nanometer']]  # Choose columns from the second DataFrame

# Concatenate the selected columns horizontally
summary_table = pd.concat([CB_den1_clean_sorted_selected, CB_den1_clean_results_df_sorted_selected], axis=1)

# Display the new DataFrame
summary_table
```
